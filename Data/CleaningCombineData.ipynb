```python
import pandas as pd
import numpy as np
```

```python
url = '/Users/VishalDubey/Vishal/College/polsci189fs/FF Project/Data/combineData0018.csv'
df = pd.read_csv(url)
df.head()
```

```python
df.describe()
```

```python
players = df[ (df['Pos'] != 'RB') & (df['Pos'] != 'TE') & (df['Pos'] != 'WR') & (df['Pos'] != 'QB') ].index

players
```

```python
df.drop(players , inplace=True)
```

```python
df.describe()
```

```python

years = df[ (df['Year'] != 2010) & (df['Year'] != 2011) & (df['Year'] != 2012) & (df['Year'] != 2013) & (df['Year'] != 2014) & (df['Year'] != 2015) & (df['Year'] != 2016) ].index

years
```

```python
df.drop(years , inplace=True)
```

```python
noTeams = df[ df['Team'].isnull() ].index
df.drop(noTeams, inplace=True)
```
