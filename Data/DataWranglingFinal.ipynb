```python
combine['Player'] = pd.Series([x.replace('-02', '') for x in combine['Player']])
combine.head()
```

```python
yr2 = pd.Series(['']*len(combine['Player']))

def createToMerge(pts, combine, year):         
    for i in range(len(pts['name'])):
        nm = pts['name'][i]
        tmp = combine['Player'][combine['Player']==nm]
        if(len(tmp)>0):
            if(combine['Year'][tmp.index[0]]==year):
                yr2[tmp.index[0]] = pts['points'][i]


yr = 2002
for csv in csvs:
    createToMerge(csv, combine, yr)
    yr+=1


#for i in range(len(yr2)):
#    print("{}\t\t{}\t\t{}".format(yr2[i], combine['Player'][i], combine['Year'][i]))
```

```python
combine['yr2'] = yr2
finalData = combine
```

```python
finalData
```

```python
count = 0
drops = []
for i in range(len(finalData.yr2)):
    if not isinstance(finalData.yr2[i], float):
        drops.append(i)
print(drops)
```

```python
finalData.drop(drops , inplace=True)
```

```python
tmp = finalData.yr2.isna()
```

```python
naDrops = []
for i in tmp.index:
    if tmp[i]:
        naDrops.append(i)
```

```python
finalData.drop(naDrops , inplace=True)
```

```python
vals = 'Ht,Wt,Forty,Vertical,BenchReps,BroadJump,Cone,Shuttle,AV,Round,Pick'.split(',')
for val in vals:
    mean = finalData[val].mean()
    nas = finalData[val].isna()
    for i in nas.index:
        if nas[i]:
            finalData[val][i] = mean
finalData
```
